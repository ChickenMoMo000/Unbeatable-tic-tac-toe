```python
import tkinter as tk
from tkinter import messagebox
import joblib
import numpy as np

class TicTacToe:
    def __init__(self, root, model_path):
        self.root = root
        self.root.title("Tic Tac Toe")

        self.board = [" " for _ in range(9)]
        self.current_player = None

        self.buttons = []
        for i in range(9):
            button = tk.Button(self.root, text=" ", font=('Arial', 20), width=4, height=2,
                               command=lambda i=i: self.make_move(i))
            button.grid(row=i // 3, column=i % 3)
            self.buttons.append(button)

        # Loading the trained Random Forest model
        self.model = joblib.load('C:\\Users\\Shahil\\your_trained_model.joblib')

        # Game start
        self.start_new_game()

    def start_new_game(self):
        self.current_player = np.random.choice(["X", "O"])
        if self.current_player == "O":
            self.make_computer_move()

    def make_move(self, index):
        if self.board[index] == " " and self.current_player == "X":
            self.board[index] = self.current_player
            self.buttons[index].config(text=self.current_player)

            if self.check_winner():
                messagebox.showinfo("Tic Tac Toe", f"Player {self.current_player} wins!")
                self.reset_board()
            elif " " not in self.board:
                messagebox.showinfo("Tic Tac Toe", "It's a draw!")
                self.reset_board()
            else:
                self.current_player = "O"
                self.make_computer_move()

    def make_computer_move(self):
        if self.current_player == "O":
            available_moves = [i for i, val in enumerate(self.board) if val == " "]

            # useing trained model to check moves
            move_probabilities = self.model_predict(self.board)
            win_or_block_move = self.check_win_or_block_move(move_probabilities)
            
            if win_or_block_move is not None:
                #make winning or blocking move 
                computer_move = win_or_block_move
            else:
                # minimax implimentationA
                computer_move = self.minimax(self.board, "O")[1]

            self.board[computer_move] = "O"
            self.buttons[computer_move].config(text="O")

            if self.check_winner():
                self.end_game("Player O wins!")
            elif " " not in self.board:
                self.end_game("It's a draw!")
            else:
                self.current_player = "X"

    def model_predict(self, board):
        # Preprocess the board state by mapping 'X' to 1, 'O' to -1, and ' ' to 0
        input_features = np.array([1 if cell == 'X' else -1 if cell == 'O' else 0 for cell in board]).reshape(1, -1)

        # Use the trained Random Forest model to predict move probabilities
        move_probabilities = self.model.predict_proba(input_features)[0]

        return move_probabilities

    def minimax(self, board, player):
        available_moves = [i for i, val in enumerate(board) if val == " "]

        if self.check_winner(board) and player == "O":
            return -1, None
        elif self.check_winner(board) and player == "X":
            return 1, None
        elif " " not in board:
            return 0, None

        moves = []

        for move in available_moves:
            new_board = board.copy()
            new_board[move] = player
            score, _ = self.minimax(new_board, "X" if player == "O" else "O")
            moves.append((score, move))

        if player == "O":
            return max(moves, key=lambda x: x[0])
        else:
            return min(moves, key=lambda x: x[0])

    def check_win_or_block_move(self, move_probabilities):
        # Check for winning or blocking moves using move probabilities
        available_moves = [i for i, val in enumerate(self.board) if val == " "]
        for move in available_moves:
            temp_board = self.board.copy()
            temp_board[move] = "O"
            if self.check_winner(temp_board):
                return move

        for move in available_moves:
            temp_board = self.board.copy()
            temp_board[move] = "X"
            if self.check_winner(temp_board):
                return move

        return None

    def check_winner(self, board=None):
        if board is None:
            board = self.board

        winning_patterns = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                            (0, 3, 6), (1, 4, 7), (2, 5, 8),
                            (0, 4, 8), (2, 4, 6)]
        for pattern in winning_patterns:
            if board[pattern[0]] == board[pattern[1]] == board[pattern[2]] != " ":
                return True
        return False

    def reset_board(self):
        self.board = [" " for _ in range(9)]
        for button in self.buttons:
            button.config(text=" ")

        self.start_new_game()

    def end_game(self, message):
        messagebox.showinfo("Tic Tac Toe", message)
        self.reset_board()

def main():
    root = tk.Tk()
    game = TicTacToe(root, 'C:\\Users\\Shahil\\your_trained_model.joblib')  # Update with your model path
    root.mainloop()

if __name__ == "__main__":
    main()
```
